```python
# Dependencies and Setup
import pandas as pd
import json
import numpy as np

# SQL Alchemy Dependencies
import sqlalchemy
from sqlalchemy import create_engine
```

```python
# Read movie_metadata.csv into movie_df
movie_df = pd.read_csv("Resources/genre_movie_swear.csv")
movie_df.rename(columns={'movie_title':'title'}, inplace=True)
```

```python
# Read all_script.csv into script_df
script_df = pd.read_csv("Resources/updated_all_script.csv")
script_df
```

```python
movie_df.dtypes
```

```python
# Select
cleaned_movie_df = movie_df
```

```python
cleaned_movie_df['profit'] = cleaned_movie_df['gross']- cleaned_movie_df['budget']
```

```python
comparison_column = np.where(cleaned_movie_df["budget"] <= cleaned_movie_df["gross"], True, False)
```

```python
cleaned_movie_df['breakeven'] = comparison_column
```

```python
cleaned_movie_df = cleaned_movie_df.drop_duplicates(subset=['title'])
```

```python
cleaned_movie_df['gross(in millions USD)'] = cleaned_movie_df['gross']/1000000
```

```python
cleaned_movie_df['budget(in millions USD)'] = cleaned_movie_df['budget']/1000000
```

```python
cleaned_movie_df['profit(in millions USD)'] = cleaned_movie_df['profit']/1000000
```

```python
cleaned_movie_df = cleaned_movie_df.drop(columns= ['gross','budget','profit'])
```

```python
movie_swear_df = pd.merge(cleaned_movie_df,script_df,how='inner',on='title')
#movie_df.join(script_df, on='title',how='outer')

movie_swear_df
```

```python
movie_swear_df = movie_swear_df.drop(columns= "Unnamed: 0" )
```

```python
movie_swear_df.to_csv("Resources/movie_swear.csv")
```
